```python
##Feature engineering process
##Splitting date of journey into date month and year and dropping the Date_of_Journey column

final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(labels='Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
##Splitting the Arrival time in to Hours and minutes and dropping the the Arrival_Time variable

final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arr_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arr_Minutes'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.drop(labels='Arrival_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arr_Hour'] = final_df['Arr_Hour'].astype(int)
final_df['Arr_Minutes'] = final_df['Arr_Minutes'].astype(int)
```

```python
final_df.info()
```

```python
##Splitting the Dept_time into dept_hour and dept_min

final_df['Dept_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.drop(labels='Dep_Time',axis=1,inplace=True)
```

```python
final_df['Dept_hour'] = final_df['Dept_hour'].astype(int)
final_df['Dept_min'] = final_df['Dept_min'].astype(int)
```

```python
final_df.info()
```

```python
##Mapping Total_Stops column for easy understanding

#Since Totsl_Stops implies the same information as Route, henceforth I dont need the Route column. So I'll drop Route


final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```
